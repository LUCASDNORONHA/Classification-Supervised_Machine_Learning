# **Arvores de Decisão**

Uma árvore de decisão é uma estrutura de dados que modela decisões e seus possíveis resultados de maneira hierárquica. Ela é uma técnica amplamente utilizada na área de ciência de dados e aprendizado de máquina.

A árvore de decisão é construída a partir de um conjunto de dados de treinamento, onde cada exemplo é representado por um conjunto de características (atributos) e uma classe ou valor alvo associado. O objetivo é criar uma árvore que represente as decisões a serem tomadas com base nas características dos dados de entrada, levando em consideração os resultados esperados.

A construção da árvore ocorre de forma recursiva, dividindo o conjunto de dados em subconjuntos cada vez menores com base em regras de decisão. Em cada nó da árvore, uma pergunta é feita com base em uma das características dos dados, e a resposta determina qual ramo seguir. Essa divisão continua até que seja alcançado um critério de parada, como uma pureza dos dados ou um limite máximo de profundidade da árvore.

Uma vez que a árvore de decisão é construída, ela pode ser utilizada para fazer previsões ou classificar novos exemplos. Ao percorrer a árvore de cima para baixo, as características do exemplo são comparadas com as condições nos nós da árvore, até chegar a uma folha que contém a classe ou valor alvo previsto.

As árvores de decisão possuem algumas vantagens, como a facilidade de interpretação e visualização, pois a estrutura da árvore pode ser facilmente compreendida. Além disso, elas podem lidar com dados numéricos e categóricos e são capazes de lidar com problemas de classificação e regressão.

No entanto, as árvores de decisão também têm algumas limitações, como a tendência a criar árvores muito complexas e propensas a overfitting (ajuste excessivo aos dados de treinamento). Para mitigar esse problema, técnicas como a poda da árvore, o uso de conjuntos de árvores (como o Random Forest) e algoritmos de regularização podem ser aplicadas.

```python
# Bibliotecas
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
from yellowbrick.classifier import ConfusionMatrix

import matplotlib.pyplot as plt
import pickle
```

# **Algoritmo de Árvore de Decisão Aplicado a Base de Crédito**



Utilizarei o modulo pickle em Python, para carregar os dados de treinamento e teste, do arquivo "credit.pkl", que criei anteriormente durante o pré-processamento dos dados.

```python
# Carregando os dados na suas respectivas variáveis.
with open('credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste  = pickle.load(f)
```

```python
X_credit_treinamento.shape, y_credit_treinamento.shape # Retorna a quantidade linhas e colunas.
```

```python
X_credit_teste.shape, y_credit_teste.shape # Retorna a quantidade linhas e colunas.
```

Com as variáveis de treinamento e teste carregadas e prontas para uso, eu poderei seguir para o processo de treinamento do algoritmo de Arvore de Decisão.

## **Treinando O Algoritmo de Árvore de Decisão**

Logo a seguir, irei criar o objeto DecisionTreeClassifier com o critério de divisão 'entropy' e um estado aleatório definido como 0.

Em seguida, treinarei o algoritmo de árvore de decisão utilizando o método fit. O modelo será ajustado aos dados de treinamento (X_credit_treinamento) e suas respectivas classes (y_credit_treinamento).

```python
# Criando o objeto DecisionTreeClassifier com criterio 'entropy'.
arvore_credit = DecisionTreeClassifier(criterion='entropy', random_state=0)

# Treinando o algoritmo de árvore de decisão utilizando o objeto arvore_credit com os dados de treinamento.
arvore_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

Algoritmo treinado e pronto para testes.

## **Testando o Algoritmo de Árvore de Decisão**

Para verificar e testar a eficácia do algoritmo, irei criar uma variável onde as previsões de classificação inferidas pelo algoritmo serão armazenadas.

Eu irei usar o modelo de árvore de decisão treinado (arvore_credit) para fazer previsões nas amostras de teste (X_credit_teste). As previsões das classes serão armazenadas na variável previsoes_credit.

```python
# Realizando as previsões das classes com amostra de teste, usando a árvore de decisão treinada anteriormente.
previsoes_credit = arvore_credit.predict(X_credit_teste)
```

Previsões armazenadas com sucesso.

**Obs.:** É importante ressaltar que o algoritmo fez suas próprias previsões e que foram armazendas na variável "previsoes_credit".

## **Avaliando O Algoritmo**

Irei verificar as previsões feitas pelo algoritmo imprimindo a variável previsoes_credit. Isso irá mostrar as previsões das classes para cada exemplo na amostra de teste feito pelo algoritmo.

```python
previsoes_credit # Retorna as previsões das classe feita pelo algoritmo.
```

A variável y_credit_teste contém as classes corretas para as amostras de teste. Ao imprimir essa variável, verei as classes reais correspondentes a cada exemplo na amostra de teste. Isso permite comparar as previsões feitas pelo algoritmo de árvore de decisão com as classes corretas e avaliar o desempenho do modelo.

```python
y_credit_teste # Retorna as classes correta.
```

A acurácia é uma métrica comum para avaliar o desempenho de um modelo de classificação. Ela mede a proporção de exemplos corretamente classificados em relação ao total de exemplos.

```python
# Calcula a acurácia do modelo de árvore de decisão ao comparar as previsões feitas pelo....
# ...modelo (previsoes_credit) com os classes verdadeiras dos exemplos de teste (y_credit_teste).
accuracy_credit =  accuracy_score(y_credit_teste, previsoes_credit)

accuracy_credit # Retorna a acurácia.
```

A biblioteca Yellowbrick fornece uma interface conveniente para visualização de métricas de aprendizado de máquina, incluindo a matriz de confusão.

```python
# Criando o objeto ConfusionMatrix.
cm = ConfusionMatrix(arvore_credit)

# Ajustando o modelo aos dados de treinamento.
cm.fit(X_credit_treinamento, y_credit_treinamento)

# Avaliando o modelo nos dados de teste.
cm.score(X_credit_teste, y_credit_teste)

# Visualizando a matriz de confusão.
cm.show();
```

Vou gerar um relatório de classificação que contém várias métricas de avaliação do modelo.

```python
# Retorna um relatório de classificação que contém várias métricas de avaliação do modelo
print(classification_report(y_credit_teste, previsoes_credit))
```

Irei visualizar a estrutura da árvore de decisão gerado pelo algoritmo.

```python
# Retorna a visualização da estrutura da árvore de decisão do algortimo.
previsores_credit = ['Income', 'age', 'loan']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(20,20))
tree.plot_tree(
    arvore_credit, feature_names=previsores_credit, class_names = ['0', '1'], filled=True
);


figura.savefig('arvore_credit.png') # O resultado será salvo em um arquivo de imagem chamado "arvore_credit.png".
```

## **Relatório Final**

Com base nas métricas fornecidas no relatório de classificação, farei a seguinte classificação do algoritmo de acordo com seu desempenho:

**Precisão:** A precisão mede a proporção de exemplos classificados corretamente em relação aos exemplos classificados como pertencentes a uma determinada classe. Neste caso, o algoritmo obteve uma precisão de 0.99 para a classe 0 e 0.91 para a classe 1. Isso indica que o algoritmo possui uma alta taxa de acertos em identificar corretamente os exemplos de ambas as classes, com uma precisão muito alta para a classe 0 e uma precisão boa para a classe 1.

**Recall:** O recall mede a proporção de exemplos corretamente identificados de uma classe em relação a todos os exemplos dessa classe. O algoritmo obteve um recall de 0.99 para a classe 0 e 0.95 para a classe 1. Isso significa que o algoritmo tem uma alta taxa de identificação correta dos exemplos das duas classes, com um recall muito alto para a classe 0 e um recall bom para a classe 1.

**F1-Score:** O F1-score é uma medida que combina a precisão e o recall, fornecendo uma métrica geral do desempenho do algoritmo. O algoritmo obteve um F1-score de 0.99 para a classe 0 e 0.93 para a classe 1. Isso indica que o algoritmo tem um bom equilíbrio entre precisão e recall para a classe 0, enquanto para a classe 1, embora ainda seja uma métrica razoável, é um pouco menor em comparação com a classe 0.

**Acurácia do algoritmo:** A acurácia mede a proporção de exemplos corretamente classificados em relação ao total de exemplos. O algoritmo obteve uma acurácia de 0.98, o que significa que ele classificou corretamente 98% dos exemplos de teste. Uma acurácia de 0.98 é considerada alta e indica um bom desempenho geral do algoritmo.

Com base nessas métricas, podemos classificar o algoritmo de árvore de decisão como tendo um bom desempenho na classificação das classes 0 e 1, com alta precisão e recall para a classe 0 e resultados razoáveis para a classe 1. A acurácia geral do algoritmo também é considerada alta.
